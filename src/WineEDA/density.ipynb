```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
df = pd.read_csv('../data/raw/winequalityN.csv')

# White wine
df = df[df['type'] == 'white']
print(df.columns[1:-1])
```

```python
df.isnull().sum()
```

```python
# fill missing values with mean
for column in df.columns[1:-1]:
    df[column].fillna(df[column].mean(), inplace=True)

df.isnull().sum()
```

```python
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(20,10))
ax = ax.flatten()

# normalize the data of each column (feature)
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[df.columns[1:-1]]), columns=df.columns[1:-1])
# concat quality column, which is not normalized
df_scaled = pd.concat([df_scaled, df['quality']], axis=1)

for i, col in enumerate(df_scaled.columns):
    if col == 'type':
        continue
    sns.kdeplot(df_scaled[col], fill=True, ax=ax[i])
    ax[i].set_title(df.columns[i+1], fontsize=14)
    
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
# set title for graph
plt.show()
```

```python
# for each quality (0~10), count how many samples are there
print(df['quality'].value_counts())
```

```python
# Define a function to plot density for a set of features on the given axes
def plot_density(ax, columns, df):
    for i, col in enumerate(columns):
        for quality in sorted(df['quality'].unique()):
            sns.kdeplot(df.loc[df['quality'] == quality, col], label=quality, ax=ax[i], fill=True, warn_singular=False)
        ax[i].set_title(col, fontsize=14)
        
        # Show the legend of quality with a title
        legend = ax[i].legend(sorted(df['quality'].unique()))
        legend.set_title("quality")

df = df_scaled.copy()

# Initialize subplots for all sets of features
figures = [plt.subplots(ncols=3, nrows=1, figsize=(20,10))[1] for _ in range(4)]

# For every 3 features, plot the density plots for each presenting quality
feature_groups = [df.columns[i:i+3] for i in range(0, len(df.columns), 3)]
for ax, features in zip(figures, feature_groups):
    plot_density(ax, features, df)

plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
plt.show()
```

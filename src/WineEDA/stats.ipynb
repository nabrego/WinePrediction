```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../../data/raw/winequalityN.csv')

df = df[df['type'] == 'white']
df = df.drop('type', axis=1)
df.head()
```

```python
# Fill missing values with the mean of each predictor column
df.fillna(df.mean(), inplace=True)

# Check for any remaining missing values
print(df.isnull().sum())
```

```python
df.describe()
```

```python
unique_scores = df["quality"].unique()
print(unique_scores)
```

```python
#Note that there are some scores that have not been given
#(df['quality'] == 2).sum()
```

```python
five_number_summaries = {}
for score in unique_scores:
    subset = df[df["quality"] == score]
    five_number_summary = subset.describe()
    five_number_summaries[score] = five_number_summary

# Display the five-number summaries
for score, summary in five_number_summaries.items():
    print(f"Simple Statistics for Target {score}:\n{summary}\n")
```

```python
#proportion of scores:
N_t = 4898
for i in range(0,11):
    proportion = (df['quality'] == i).sum()/N_t
    print(f"Proportion of quality score {i}: {proportion :.3f}")
```

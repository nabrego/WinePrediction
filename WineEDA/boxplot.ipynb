```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/winequalityN.csv')

# White wine
dfW = df[df['type'] == 'white']
dfW = dfW.drop('type', axis=1)

# Red wine
dfR = df[df['type'] == 'red']
dfR = dfR.drop('type', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

# We standardize the data so we can see better every attribute
scaler = StandardScaler()
dfW_scaled = pd.DataFrame(scaler.fit_transform(dfW), columns=dfW.columns)
dfR_scaled = pd.DataFrame(scaler.fit_transform(dfR), columns=dfR.columns)


# White wine boxplot with labels
plt.boxplot(dfW_scaled.values)
plt.xticks(range(1, len(dfW_scaled.columns) + 1), dfW_scaled.columns, rotation=90)
plt.title('White Standardized Wine')
plt.show()


# Red wine boxplot with labels
plt.boxplot(dfR_scaled.values)
plt.xticks(range(1, len(dfR_scaled.columns) + 1), dfR_scaled.columns, rotation=90)
plt.title('Red Standardized Wine')
plt.show()
```
